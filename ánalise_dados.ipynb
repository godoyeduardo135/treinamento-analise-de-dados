```python
salarios = np.array([2,3,4,5,6,700])
maior_salarios = np.max(salarios)
print(maior_salarios)
```

```python
medias_salarios = np.mean(salarios)
print(f"{medias_salarios:.2f}")
```

```python
minimo_salario = np.min(salarios)
print(minimo_salario)
```

```python
mediana_salarios = np.median(salarios)
print(mediana_salarios)
```

```python
desvio = np.std(salarios)
print(desvio)
```

```python
Criar uma lista de venda por dia 
-Vendas = [10,20,45,5,90]
-Exibir a maior venda 
-Exibir a menor venda 
-Exibir a média de vendas
-Exibir a mediana de vendas
```

```python
Venda_dia = np.array ([10,20,45,5,90])
maior_salarios = np.max(Venda_dia)
print(maior_salarios)

medias_vendas = np.mean(Venda_dia)
print(f"{medias_vendas:.2f}")


minimo_vendas = np.min(Venda_dia)
print(minimo_vendas)

media_vendas = np.mean(Venda_dia)
print(media_vendas)

mediana_vendas = np.median(Venda_dia)
print(mediana_vendas)

desvio_venda = np.std(Venda_dia)
print(desvio_venda)
```
