```python
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv(r".\data\dataset_aula01_techshop_vendas.csv")
df
```

```python
grafico = sns.countplot(data=df,x="categoria")
grafico.set_title("contagem por categoria")
grafico.set_ylabel("Quantidade");
grafico.set_xlabel("Categoria");
```

```python
grafico = sns.histplot(data=df,x="categoria",y="preco_unitario",kde=True,bins=20)
grafico.set_title("Histograma");
grafico.set_ylabel("Quantidade de vendas");
grafico.set_xlabel("Preços dos produtos");
```

```python
graico = sns.boxplot(data=df,x="categoria",y="preco_unitario")
grafico.set_title("Distrubuição de preços por categorias");
grafico.set_ylabel("Categoria");
grafico.set_xlabel("Preços");
```
